```python
import gym
```

```python
env = gym.make('CartPole-v0')

for i_episode in range(20):
    
    # This resets the environment's state to the initial default conditions. 
    # It actually returns the start observations.
    observation = env.reset() # 
    
    for t in range(100):
        env.render()
        print(observation)
        
        # taking a random action
        action = env.action_space.sample()
        
        #  This moves the environment by a single time step. 
        # It returns a four-valued vector made of variables: observations, reward, done, and info
        observation, reward, done, info = env.step(action)
        
        if done:
         print("Episode finished after %i timesteps" % (t+1))
         break
```
